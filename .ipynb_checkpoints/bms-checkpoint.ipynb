<h3>Defining Exception Classes</h3>

```python
class Error(Exception):
    def __init__(self, msg):
        self.msg = msg
        super().__init__(self)

    def printmsg(self):
        input(self.msg)


class TerminateProgramException(Error):
    def __init__(self, msg="Bye! Hope you run this program again\nHit <enter_key> to continue ..."):
        super().__init__(msg)


class NoNumException(Error):
    def __init__(self, msg="NoNumException : Enter ENGLISH ALPHABETS ONLY!\nHit <enter_key> to continue ..."):
        super().__init__(msg)


class NoAlphaException(Error):
    def __init__(self, msg="NoAlphaException : Enter NUMBERS ONLY!\nHit <enter_key> to continue ..."):
        super().__init__(msg)


class PassMismatchException(Error):
    def __init__(self, msg="PassMismatchException : Passwords DO NOT match!\nHit <enter_key> to continue ..."):
        super().__init__(msg)


class InvalidCredsException(Error):
    def __init__(self, msg="InvalidCredsException : Invalid Credentials have been entered!\nCheck credentials and Try Again!!\nHit <enter_key> to continue ..."):
        super().__init__(msg)


class NoSignInException(Error):
    def __init__(self, msg="NoSignInException : Sign-In Required for the desired operation\nSign-In and Try Again!!\nHit <enter_key> to continue ..."):
        super().__init__(msg)


class LowBalException(Error):
    def __init__(self, msg="LowBalException : Withdrawal NOT possible!!\nHit <enter_key> to continue ..."):
        super().__init__(msg)


class AccountThresholdException(Error):
    def __init__(self, msg="AccountThresholdException : Deposit NOT possible!!\nAccount threshold reached!!\nHit <enter_key> to continue ..."):
        super().__init__(msg)
```

<h3>Defining Classes</h3>

```python
class Customer:
    def __init__(self, custid, name, phone, age, password, initDepo):
        # self.balance = initDepo
        self.transactions = []
        self.custInfo = {
            "custid": custid,
            "name": name,
            "phone": phone,
            "age": age,
            "password": password,
            "balance": initDepo
            # "balance": self.balance
        }

    def chkBal(self):
        print('\nYour Customer ID : ', self.custInfo["custid"])
        print(self.custInfo["name"], ', you have a balance of : ₹',
              self.custInfo["balance"], '/-')

    def withDraw(self, cash):
        if (self.custInfo["balance"] - 1000) < cash:
            raise LowBalException
        self.custInfo["balance"] -= cash
        self.transactions.append(-abs(cash))
        print('₹ ', cash, '/- has been WITHDRAWN FROM your account successfully!!')
        # \nAvailable Balance : ₹',self.custInfo["balance"], '/-')
        self.chkBal()

    def deposit(self, cash):
        if cash >= 200000:
            raise AccountThresholdException
        self.transactions.append(cash)
        self.custInfo["balance"] += cash
        print('₹ ', cash, '/- has been DEPOSITED TO your account successfully!!')
        # \nAvailable Balance : ₹',self.custInfo["balance"], '/-')
        self.chkBal()

    # def last5(self):
    #     print('Details of Last 5 Transactions from customer ID : ',
    #           self.custInfo["custid"])
    #     while(len(self.transactions) < 5):
    #         for i in range(0, len(self.transactions)):
    #             if self.transactions[len(self.transactions)-i-1] > 0:
    #                 print(6-i, '.) DEPOSITED : ₹',
    #                       self.transactions[:-i], '/-')
    #             else:
    #                 print(6-i, '.) WITHDRAWN : ₹',
    #                       self.transactions[:-i], '/-')

    def last5(self):
        print('\nDetails of all Transactions for customer ID : ', self.custInfo["custid"])
        self.chkBal()
        if(len(self.transactions) >= 5):
            self.transactions.reverse()
            for x in range(0, 5):
                if self.transactions[x] > 0:
                    print('\n(', x+1, '). DEPOSITED : ₹',
                          self.transactions[x], '/-')
                else:
                    print('\n(', x+1, '). WITHDRAWN : - ₹',
                          self.transactions[x]*-1, '/-')
            self.transactions.reverse()
            #print('\nAvailable Balance : ', self.custInfo["balance"])
        else:
            print('\nNot enough transactions to be printed!!')

    def delAcc(self):
        print('\nYou are about to delete the account with Customer ID : ',
              self.custInfo["custid"])
        passw = input('\nRe-enter password to delete the account : ')
        if passw == self.custInfo["password"]:
            return 1
        else:
            raise InvalidCredsException
```

<h3>Defining global functions</h3>

```python
def menu():
    print('[ 1 ] : OPEN ACCOUNT')
    print('[ 2 ] : SIGN IN')
    print('[ 3 ] : SIGN OUT')
    print('[ 4 ] : CHECK BALANCE')
    print('[ 5 ] : WITHDRAW CASH')
    print('[ 6 ] : DEPOSIT CASH')
    print('[ 7 ] : CHECK LAST 5 TRANSACTIONS')
    print('[ 8 ] : DELETE ACCOUNT')
    print('[ L ] : LOAD CUSTOMER INFORMATION')
    print('[ S ] : SAVE & EXPORT CUSTOMER INFORMATION')
    print('[ 0 ] : EXIT')
    choice = input('[ > ] : ')[0].upper()
    # if(re.findall("[^0-9]", choice)):
    #     raise NoAlphaException
    return choice


def openAcc(custid):
    # print('\nEnter PAN Number : ')
    # print('\nEnter Aadhaar Number : ')
    name = input('\nEnter your name : ')
    if(re.findall("[^a-z^A-Z]", name)):
        raise NoNumException
    age = input('Enter your age : ')
    if(re.findall("[^0-9]", age)):
        raise NoAlphaException
    phn = input('Enter your phone number : ')
    if(re.findall("[^0-9]", phn)):
        raise NoAlphaException
    initDepo = input('Enter your initial deposit : ')
    if(re.findall("[^0-9]", phn)):
        raise NoAlphaException
    passw = input('\nEnter password :')
    cpass = input('Confirm password :')

    if(passw == cpass):
        return Customer(int(custid), name, int(phn), int(age), passw, int(initDepo))
    else:
        raise PassMismatchException


def signIn(customers, phn, passw):
    for x in range(0, len(customers)):
        # for x in enumerate(customers, 0):
        if customers[x].custInfo["phone"] == phn and customers[x].custInfo["password"] == passw:
            return x
    return -1
```

<h3>Driver Code</h3>

```python
if __name__ == '__main__':
    customers = []
    current = -1
    lengTH = 0
    seedId = 90001212900
    while True:
        os.system('cls')
        try:
            choice = menu()

            # OPEN ACCOUNT
            if choice == '1':
                customers.append(openAcc(seedId))
                print('\nYour entries have been recorded successfully!\nYour new Customer ID is : ',
                      customers[lengTH].custInfo["custid"])
                lengTH += 1
                seedId += 1
            # SIGN-IN
            elif choice == '2':
                phn = input('\nEnter phone number : ')
                if(re.findall("[^0-9]", phn)):
                    raise NoAlphaException
                passw = input('Enter password     : ')
                current = signIn(customers, int(phn), passw)
                if(current == -1):
                    raise InvalidCredsException
                else:
                    print('Signed in Successfully!\nYour Customer ID is : ',
                          customers[current].custInfo["custid"])
            # SIGN-OUT
            elif choice == '3':
                current = -1
                print('\nSigned Out Successfully!!')
            # CHECK BALANCE
            elif choice == '4':
                if(current == -1):
                    raise NoSignInException
                else:
                    customers[current].chkBal()
            # WITHDRAW CASH
            elif choice == '5':
                if(current == -1):
                    raise NoSignInException
                else:
                    cash = input('\nEnter amount to withdraw : ')
                    if(re.findall("[^0-9]", cash)):
                        raise NoAlphaException
                    customers[current].withDraw(int(cash))
            # DEPOSIT CASH
            elif choice == '6':
                if(current == -1):
                    raise NoSignInException
                else:
                    cash = input('\nEnter amount to deposit : ')
                    if(re.findall("[^0-9]", cash)):
                        raise NoAlphaException
                    customers[current].deposit(int(cash))
            # CHECK LAST 5 TRANSACTIONS
            elif choice == '7':
                if(current == -1):
                    raise NoSignInException
                customers[current].last5()
            # DELETE ACCOUNT
            elif choice == '8':
                if(current == -1):
                    raise NoSignInException
                if customers[current].delAcc():
                    cid = customers[current].custInfo["custid"]
                    del customers[current]
                    lengTH -= 1
                    current = -1
                    print('\nSigned out successfully!!\nAccount with customer ID : ',
                          cid, ' has been deleted permanently!!')
            # LOAD CUSTOMER INFORMATION
            elif choice == 'L':
                print('\nopening file : customers.txt ....')
                with open('customers.txt', 'rt') as toLoad:
                    print('\nreading from file : customers.txt ....')
                    for row in toLoad:
                        temp = ''
                        count = 0
                        for char in row:
                            if char == ' ':
                                continue
                            if char == ';':
                                balance = temp
                                break
                            if char != ',':
                                temp += char
                            if char == ',':
                                count += 1
                                if count == 1:
                                    custid = temp
                                elif count == 2:
                                    name = temp
                                elif count == 3:
                                    phn = temp
                                elif count == 4:
                                    age = temp
                                elif count == 5:
                                    passw = temp
                                else:
                                    print(
                                        '\nCorrupted Data!!\nError Reading from file!!')
                                temp = ''
                        customers.append(Customer(int(custid), name, int(
                            phn), int(age), passw, int(balance)))
                        lengTH += 1
                print('\nclosing file : customers.txt ....')
                toLoad.close()
                print('\nAll Records have been loaded successfully!!')
            # SAVE & EXPORT CUSTOMER INFORMATION
            elif choice == 'S':
                print('\nopening file : customers.txt ....')
                with open('customers.txt', 'w') as toWrite:
                    print('\nwriting to file : cutomers.txt ....')
                    for x in range(len(customers)):
                        txt = str(customers[x].custInfo["custid"]) + ',' + customers[x].custInfo["name"] + ',' + str(customers[x].custInfo["phone"]) + ',' + str(
                            customers[x].custInfo["age"]) + ',' + customers[x].custInfo["password"] + ',' + str(customers[x].custInfo["balance"]) + ';\n'
                        print(txt)
                        toWrite.write(txt)
                print('\nclosing file : customers.txt ....')
                toWrite.close()
                print('\nAll Records have been saved and exported successfully!!')
            # EXIT
            elif choice == '0':
                raise TerminateProgramException
            else:
                print('\nINVALID INPUT\nTRY AGAIN!!')

            input('\nHit <enter_key> to continue ...')

        # EXCEPTIONS
        except TerminateProgramException as TPE:
            TPE.printmsg()
            quit()

        except NoNumException as NNE:
            NNE.printmsg()
            continue

        except NoAlphaException as NAE:
            NAE.printmsg()
            continue

        except PassMismatchException as PME:
            PME.printmsg()
            continue

        except InvalidCredsException as ICE:
            ICE.printmsg()
            continue

        except NoSignInException as NSIE:
            NSIE.printmsg()
            continue

        except LowBalException as LBE:
            LBE.printmsg()
            continue

        except AccountThresholdException as ATE:
            ATE.printmsg()
            continue
```
